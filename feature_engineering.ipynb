### 1. Merge All Relevant Data

```python
# Merge transactions with courses
data = transactions.merge(courses, on='CourseID', how='left')
data
```

```python
# Merge with teachers (assuming TeacherID exists in courses)
data = data.merge(teachers, on='TeacherID', how='left')
data
```

```python
data.head()
```

### 2. Engagement Features

#### . Total Courses Entrolled

```python
total_courses = data.groupby('UserID')['CourseID'].count().reset_index()
total_courses.rename(columns={'CourseID': 'total_courses'}, inplace=True)
```

```python
total_courses
```

#### . Enrollment Frequency

```python
data['TransactionDate'] = pd.to_datetime(data['TransactionDate'])

enrollment_freq = data.groupby('UserID')['TransactionDate'].count().reset_index()
enrollment_freq.rename(columns={'TransactionDate': 'enrollment_frequency'}, inplace=True)
```

```python
enrollment_freq
```

#### . Courses per Category

```python
courses_per_category = data.groupby(['UserID', 'CourseCategory']).size().unstack(fill_value=0)
courses_per_category.reset_index(inplace=True)
```

```python
courses_per_category
```

### 3. Preference Features

#### . Preferred Category

```python
pref_category = data.groupby(['UserID', 'CourseCategory']).size().reset_index(name='count')

pref_category = pref_category.sort_values(['UserID', 'count'], ascending=False)
pref_category = pref_category.drop_duplicates('UserID')

pref_category = pref_category[['UserID', 'CourseCategory']]
pref_category.rename(columns={'CourseCategory': 'preferred_category'}, inplace=True)
```

```python
pref_category
```

#### . Preferred Course level

```python
pref_level = data.groupby(['UserID', 'CourseLevel']).size().reset_index(name='count')

pref_level = pref_level.sort_values(['UserID', 'count'], ascending=False)
pref_level = pref_level.drop_duplicates('UserID')

pref_level = pref_level[['UserID', 'CourseLevel']]
pref_level.rename(columns={'CourseLevel': 'preferred_level'}, inplace=True)
```

```python
pref_level
```

#### . Average Course Rating

```python
avg_rating = data.groupby('UserID')['CourseRating'].mean().reset_index()
avg_rating.rename(columns={'CourseRating': 'avg_course_rating'}, inplace=True)
```

```python
avg_rating
```

### 4. Behavioral Features

#### . Average Spending

```python
avg_spending = data.groupby('UserID')['Amount'].mean().reset_index()
avg_spending.rename(columns={'Amount': 'avg_spending'}, inplace=True)
```

```python
avg_spending
```

#### . Total Spending

```python
total_spending = data.groupby('UserID')['Amount'].sum().reset_index()
total_spending.rename(columns={'Amount': 'total_spending'}, inplace=True)
```

```python
total_spending
```

#### . Diversity Score (Categories Explored)

```python
diversity = data.groupby('UserID')['CourseCategory'].nunique().reset_index()
diversity.rename(columns={'CourseCategory': 'diversity_score'}, inplace=True)
```

```python
diversity
```

#### . Learning Depth Index (Beginner vs Advanced)

```python
level_counts = data.groupby(['UserID', 'CourseLevel']).size().unstack(fill_value=0)

# Avoid error if columns missing
if 'Beginner' not in level_counts:
    level_counts['Beginner'] = 0
if 'Advanced' not in level_counts:
    level_counts['Advanced'] = 0

level_counts['learning_depth_index'] = level_counts['Advanced'] / (level_counts['Beginner'] + 1)

level_counts = level_counts[['learning_depth_index']].reset_index()
```

```python
level_counts
```

### 5. Teacher-Based Features

#### . Preferred Teacher

```python
pref_teacher = data.groupby(['UserID', 'TeacherID']).size().reset_index(name='count')

pref_teacher = pref_teacher.sort_values(['UserID', 'count'], ascending=False)
pref_teacher = pref_teacher.drop_duplicates('UserID')

pref_teacher = pref_teacher[['UserID', 'TeacherID']]
pref_teacher.rename(columns={'TeacherID': 'preferred_teacher'}, inplace=True)
```

```python
pref_teacher
```

#### . Teacher Diversity

```python
teacher_diversity = data.groupby('UserID')['TeacherID'].nunique().reset_index()
teacher_diversity.rename(columns={'TeacherID': 'teacher_diversity'}, inplace=True)
```

```python
teacher_diversity
```

### 6. Final Learner Profile

```python
learner_profile = users.copy()

dfs = [
    total_courses,
    enrollment_freq,
    avg_spending,
    total_spending,
    diversity,
    avg_rating,
    pref_category,
    pref_level,
    level_counts,
    pref_teacher,
    teacher_diversity
]

for df in dfs:
    learner_profile = learner_profile.merge(df, on='UserID', how='left')

# Fill missing values
learner_profile.fillna(0, inplace=True)

learner_profile.head()
```

### 7. Save Final Dataset

```python
learner_profile.to_csv("../outputs/learner_profile.csv", index=False)
```
